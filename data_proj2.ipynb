```python
import pandas as pd
from sqlalchemy import create_engine
```

```python
premise_file = "data_min_vers.csv"
data_min_df = pd.read_csv(premise_file)
data_min_df.head()
```

```python
data_cols = ["flight_number", "launch_date_utc", "launch_site/site_name", "rocket/first_stage/cores/0/core_serial", "rocket/second_stage/payloads/0/payload_id",
            "rocket/second_stage/payloads/1/payload_mass_kg", "rocket/second_stage/payloads/1/orbit",
             "rocket/second_stage/payloads/0/customers/0", "launch_success", "rocket/first_stage/cores/0/land_success"]
data_min_df_transformed= data_min_df[data_cols].copy()
data_min_df_transformed
```

```python
data_min_1 = data_min_df_transformed.rename(columns={"flight_number": "flight_Number",
                                          "launch_date_utc": "launch_date",
                                          "rocket/first_stage/cores/0/core_serial": "booster_version",
                                          "launch_site/site_name": "launch_site",
                                          "rocket/second_stage/payloads/0/payload_id":"payload_name",
                                          "rocket/second_stage/payloads/1/payload_mass_kg":"payload_mass",
                                          "rocket/second_stage/payloads/1/orbit":"payload_orbit",
                                          "rocket/second_stage/payloads/0/customers/0":"customer_name",
                                          "launch_success":"mission_outcome",
                                          "rocket/first_stage/cores/0/land_success":"landing_outcome"})
data_min_1
```

```python
data_min_1.to_csv("data_min_revised.csv", encoding='utf-8', index=False)
```
